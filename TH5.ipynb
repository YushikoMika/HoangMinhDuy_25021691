```python
import math
def is_prime_number(n):
  if n <= 1:
    return False
  for i in range (2,math.isqrt(n)+1):
    if n%i == 0:
      return False
    else:
      return True
n = int(input())
print(is_prime_number(n))
```

```python
def perferct_num(n):
  sum = 0
  for i in range (1,n+1):
    if n%i == 0:
      sum += 1
  if sum == n:
    return True
  else:
    return False
n = int(input())
print(perferct_num(n))
```

```python
def k_in_list(k,t):
  if k in t:
    return t.index(k)
  else:
    return -1
k = int(input())
t = list(map(int,input().split()))
print(k_in_list(k,t))
```

```python
def giai_thua_n(n):
  prod = 1
  for i in range (1,n+1):
    prod *= i
  return prod
n = int(input())
print(giai_thua_n(n))
```

```python
def calc(num1:float,num2:float,operat:str):
  result = 0.0
  if operat == '+':
    result = num1 + num2
  elif operat == '-':
    result = num1 - num2
  elif operat == '*':
    result = num1 * num2
  elif operat == '/':
    if num2 == 0:
      return False
    result = num1/num2
  else:
    print("Error")
  return round(result,2)
num1, num2 = map(float,input().split())
operat = input()
print(calc(num1,num2,operat))
```

```python
def hamming_distance_int(n1,n2):
  xor_result = n1 ^ n2
  binary_string = bin(xor_result)[2:]
  return binary_string.count('1')
n1, n2 = map(int,input().split())
print(hamming_distance_int(n1,n2))
```

```python
def sum_each_digit(n):
  absolute_num = abs(n)
  if absolute_num == 0:
    return False
  digits= []
  while absolute_num > 0:
    last_digit = absolute_num % 10
    digits.append(last_digit)
    absolute_num //= 10
  return sum(digits)
n = int(input())
print(sum_each_digit(n))
```

```python
def is_isomorphic(a:str,b:str):
  map_a_to_b = {}
  map_b_to_a = {}
  n = len(a)
  for i in range (n):
    char_a = a[i]
    char_b = b[i]
    if char_a in map_a_to_b:
      if map_a_to_b[char_a] != char_b:
        return False
    else:
      map_a_to_b[char_a] = char_b
    if char_b in map_b_to_a:
      if map_b_to_a[char_b] != char_a:
        return False
    else:
      map_b_to_a[char_b] = char_a
  return True
a,b = map(str,input().split())
print(is_isomorphic(a,b))
```
